```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind, normaltest, ranksums

gender_data = pd.read_csv('C:/Users/cphil/Documents/GitHub/MATH495/DBSF_removedDiff300.csv')
female = gender_data[gender_data['S1'] == 1]
male = gender_data[gender_data['S1'] == 2]
sns.scatterplot(x = 'USERID', y = 'A2', data = female)
sns.scatterplot(x = 'USERID', y = 'A2', data = male)
plt.show()
```

This is a scatterplot to show all calories intook by each user (only includes males and females)

Next we will perform a rank-sum to find if the population means are equal, to find if the average calories for the population are equal:

H0: μ1 = μ2
HA: μ1 =/= μ2

```python
print(normaltest(female['A2']))
print(normaltest(male['A2']))
```

Both samples fail the normality test, so we will use the Rank-Sum test instead of 2-Sample T-Test


```python
print(ttest_ind(female['A2'], male['A2']))
```

P Value < .05, therefore we cannot confirm the hypothesis that the average population calories is equal.
